```python
#Showing the % of players drafted per position
plt.figure(figsize=(14, 14));
nfl_positions = nfl['Position Standard'].value_counts();
nfl_positions.plot(kind='pie', autopct='%1.0f%%', fontsize=13);
plt.title('Percentage of Players Drafted per Position 1985 - 2015', fontsize=20);
```

```python
plt.figure(figsize=(20, 7))
nfl[nfl.Rnd <= 6]['College/Univ'].value_counts()[:15].plot(kind='bar', color='purple');

plt.xlabel('College or University', size=15);
plt.ylabel('# players drafted', size=15)
plt.title('Top Colleges/Univ with Players Drafted', size=20);
```

```python

nfl.loc[nfl['PB']>0,['Player','Year']].groupby(['Year']).count().sort_values('Player', ascending = False).plot(kind='bar', figsize=(18, 7), color='purple')
plt.xlabel('Draft Year', size=14)
plt.ylabel('# of players selected for pro bowl', size=14)
plt.title('Best draft class based on pro bowl selection', size=20);

```

```python
draft_class = nfl.loc[nfl['PB']>0,['Player','Year']].groupby(['Year']).count() / nfl[['Player','Year']].groupby(['Year']).count()
draft_class.sort_values('Player', ascending = False, inplace=True)
```

```python
draft_class.plot(kind='bar', figsize=(18, 7), color = 'purple');
plt.title('Average player per drafted selected for Pro Bowl', size=20);
```

```python
nfl.loc[nfl['PB']>0,['Player','College/Univ']].groupby(['College/Univ']).count().sort_values('Player', ascending=False)[:20].plot(kind='bar', figsize=(18, 7), color ='purple')
plt.xlabel('College/University', size=14);
plt.ylabel('# of player selected', size=14);
plt.title('College/Univ with most drafted pro bolwers', size =20);

```

```python
#plt.figure(figsize=(20, 7));
nfl.loc[nfl['PB']>0,['Player','Tm']].groupby(['Tm']).count().sort_values('Player', ascending=True)[25:].plot(kind="barh", figsize=(18, 5), color = 'black')
plt.xlabel("# Players select to Pro Bowl", size = 14);
plt.ylabel('Teams', size = 14);
plt.title('Pro bowlers per Team draft by', size=20);
```

```python
drafted_qb = nfl.loc[nfl.Pos == 'QB']
drafted_qb.groupby(['College/Univ']).count().sort_values('Player', ascending=False)[:10]

```

```python
nfl_pivot = nfl.pivot_table(values='Player', 
                            index=['College/Univ'], 
                            columns=['Pos'], 
                            aggfunc='count')

my_colors = ['r', 'orange','yellow', 'g','b','purple']*10
```

```python
nfl_pivot['RB'].sort_values(ascending=False)[:20].plot(kind="bar", figsize=(18, 5), color = my_colors);
plt.title('# of Drafted Running Backs per College/University', size= 20);
```

```python
nfl_pivot['DB'].sort_values(ascending=False)[:20].plot(kind="bar", figsize=(18, 5), color = my_colors);
plt.title('# of Drafted DB per College/University', size= 20);
```
